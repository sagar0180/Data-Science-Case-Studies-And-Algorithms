```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
wine = pd.read_csv("D:/DATA SCIENCE/1. Data Scientist/Assignment/7. PCA/wine.csv")
```

```python
wine.head()
```

```python
# considering valid /important column or data
wine1 = wine.iloc[:,1:]
wine1.head()
```

```python
# convert it into numpy array
wine2 = wine1.values
wine2
```

```python
# like z-score normalize the data
#scale data
wine3 = scale(wine2)
wine3
```

```python
# main working code in pca
pca = PCA()
pca_values = pca.fit_transform(wine3)
pca_values
```

```python
# using first 3 principal component scores
pca = PCA(n_components = 3)
pca_values = pca.fit_transform(wine3)
pca_values
```

```python
#the amount of variation that explained by each pca
var = pca.explained_variance_ratio_
var
```

```python
# cumulative variance
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
pca.components_
```

```python
# var plot for pca component obtained
plt.plot(var1,color='red')
```

```python
# plot between pca1,pca2,pca3
x = pca_values[:,0:1]
y = pca_values[:,1:2]
z = pca_values[:,2:3]
plt.scatter(x,y,z)
```

```python
finaldf = pd.concat([pd.DataFrame(pca_values[:,0:3],columns=['pc1','pc2','pc3']), wine[['Type']]], axis=1)
```

```python
# using elbow method
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(wine3)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
model = KMeans(n_clusters=3)
model.fit(wine3)
model.labels_
```

```python
md = pd.Series(model.labels_)
wine['clust'] = md
wine
```

```python
wine.iloc[:,1:13].groupby(wine.clust).mean()
```
