```python
data_new=data.drop(columns=['Unnamed: 0'])
```

```python
#Remember that pivot is a dead simple data reshape and it doesn’t do any aggregation. So when we are trying to pivot our data frame, our index column (which goes to the index) and columns column (which goes to the header) combination must not have any duplicates, otherwise, pivot doesn’t know how to reshape your data.
```

```python
user_data = data_new.pivot_table(index='User_ID',
                           columns='Book_Title',
                           values='Book_Rating',aggfunc='sum')
```

```python
df_agg = data_new.groupby(by=['User_ID', 'Book_Title']).Book_Rating.sum().reset_index()
```

```python
dt = df_agg.pivot(index='User_ID', columns='Book_Title', values='Book_Rating')
```

```python
dt
```

```python
dt.index = data.User_ID.unique()
```

```python
dt
```

```python
# impute tose NANS with value 0
dt.fillna(0,inplace=True)
```

```python
dt
```

```python
#calculating cosine similarity between user
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( dt.values,metric='cosine')
```

```python
user_sim
```

```python
# store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
# set the index and column names to user ids
user_sim_df.index = data.User_ID.unique()
user_sim_df.columns = data.User_ID.unique()
```

```python
user_sim_df.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```
