```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
groceries = pd.read_csv("D:/DATA SCIENCE/1. Data Scientist/Assignment/8. association rule/groceries.csv")
```

```python
groceries.head()
```

```python
groceries.shape
```

```python
df = pd.get_dummies(groceries)
df.head()
# here already we have segregated data so no use of getting dummies variables
```

```python
frequent_itemsets = apriori(df,min_support=0.0022, use_colnames=True)
frequent_itemsets
#Min_support  = 3(3 times a day) * 7 (7 days a week) / 9834 = 0.0022
```

```python
rules = association_rules(frequent_itemsets,metric='lift',min_threshold = 0.7)
rules
```

```python
rules.sort_values('lift',ascending = False)[0:20]
```

```python
rules[rules.lift>1]
```
