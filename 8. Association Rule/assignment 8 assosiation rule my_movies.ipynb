```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
my_movies = pd.read_csv("D:/DATA SCIENCE/1. Data Scientist/Assignment/8. association rule/my_movies.csv")
```

```python
my_movies.head()
```

```python
my_movies.info()
```

```python
df1 = pd.get_dummies(my_movies)
df1.head()
```

```python
frequent_itemsets1 = apriori(df1,min_support=0.1, use_colnames=True)
frequent_itemsets1
```

```python
rules1 = association_rules(frequent_itemsets1,metric='lift',min_threshold = 0.7)
rules1
```

```python
rules1.sort_values('lift',ascending = False)[0:20]
```

```python
rules1[rules1.lift>1]
```
