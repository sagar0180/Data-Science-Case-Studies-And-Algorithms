```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
book = pd.read_csv("D:/DATA SCIENCE/1. Data Scientist/Assignment/8. association rule/book.csv")
```

```python
book.head()
```

```python
book.info()
```

```python
df = pd.get_dummies(book)
df.head()
# here already we have segregated data so no use of getting dummies variables
```

```python
# aproiri algorithm
```

```python
frequent_itemsets = apriori(df,min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets,metric='lift',min_threshold = 0.7)
rules
```

```python
rules.sort_values('lift',ascending = False)[0:20]
```

```python
rules[rules.lift>1]
```
