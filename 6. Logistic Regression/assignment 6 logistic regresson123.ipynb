```python
# DROPING THE CASE NUMBER IT IS NOT REQUIRED
bank.drop(["contact","poutcome"],inplace=True,axis=1)
```

```python
from sklearn import preprocessing
Le= preprocessing.LabelEncoder()
bank['job_1']=Le.fit_transform(bank['job'])
bank['marital_1']=Le.fit_transform(bank['marital'])
bank['education_1']=Le.fit_transform(bank['education'])
bank['default_1']=Le.fit_transform(bank['default'])
bank['housing_1']=Le.fit_transform(bank['housing'])
bank['loan_1']=Le.fit_transform(bank['loan'])
bank['month_1']=Le.fit_transform(bank['month'])
```

```python
bank.head()
```

```python
bank = bank.drop(['job','marital','education','default','housing','loan','month'],axis=1)
```

```python
bank = bank[['y','age', 'balance', 'day', 'duration', 'campaign', 'pdays', 'previous',
        'job_1', 'marital_1', 'education_1', 'default_1', 'housing_1',
       'loan_1', 'month_1']]
```

```python
bank = bank.dropna()
bank.shape
```
